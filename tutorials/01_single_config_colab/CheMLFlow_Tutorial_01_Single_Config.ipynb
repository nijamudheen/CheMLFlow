## 3. Inspect the config

This is a single runnable slice from a 5-fold CV design. Load it as a Python object instead of dumping raw file text.

```python
from pathlib import Path
import yaml

CONFIG_PATH = Path("tutorials/01_single_config_colab/configs/pgp_svm_cv_fold0.yaml")
config = yaml.safe_load(CONFIG_PATH.read_text())
print(yaml.safe_dump(config, sort_keys=False))
```

## 4. Run CheMLFlow

This executes fold 0 of repeat 0 from the 5-fold CV setup.

```python
import os
import subprocess
import sys

os.environ["CHEMLFLOW_CONFIG"] = str(CONFIG_PATH)
result = subprocess.run([sys.executable, "main.py"], text=True, capture_output=True)
print(result.stdout)
if result.stderr:
    print(result.stderr)
result.check_returncode()
```

## 5. Read the output metrics

```python
from pathlib import Path
import json

run_dir = Path("tutorials/01_single_config_colab/artifacts/runs/pgp_svm_cv_fold0")
metrics = json.loads((run_dir / "svm_metrics.json").read_text())
split_metrics = json.loads((run_dir / "svm_split_metrics.json").read_text())

print("Run dir:", run_dir)
print("\nPrimary metrics:")
print(json.dumps(metrics, indent=2))
print("\nSplit metrics:")
print(json.dumps(split_metrics, indent=2))
```
